### GROUP BY

- O comando `GROUP BY` agrupa linhas baseado em semelhan√ßas sentre elas
- Por exemplot, podemos contar o n√∫mero de produtos  em cada categoria com a fun√ß√£o `COUNT`

https://www.devmedia.com.br/desvendando-a-clausula-group-by-artigo-sql-magazine-47/8082

```python
query = '''
    SELECT *
    FROM dbo.olist_products_dataset
    WHERE product_category_name = 'artes'
'''

pd.read_sql(query, engine)
```

```python
query = '''
    SELECT product_category_name, COUNT(*) AS Num_produtos
    FROM dbo.olist_products_dataset
    GROUP BY product_category_name
'''

pd.read_sql(query, engine)
```

```python
query = '''
    SELECT product_category_name
            ,COUNT(*) AS Num_produtos
            ,AVG(product_weight_g) AS Peso_medio
    FROM dbo.olist_products_dataset
    GROUP BY product_category_name
'''

pd.read_sql(query, engine)
```

### ORDER BY

- O comando `ORDER BY` orderna os valores por determinada colunas

https://www.devmedia.com.br/sql-order-by/41225

```python
query = '''
    SELECT product_category_name, COUNT(*) AS Num_produtos
    FROM dbo.olist_products_dataset
    GROUP BY product_category_name
    ORDER BY Num_produtos
'''

pd.read_sql(query, engine)
```

Podemos ordenar em valores decrescentes

```python
query = '''
    SELECT product_category_name, COUNT(*) AS Num_produtos
    FROM dbo.olist_products_dataset
    WHERE product_category_name IN ('esporte_lazer', 'utilidades_domesticas')
    GROUP BY product_category_name
    ORDER BY Num_produtos DESC
'''

pd.read_sql(query, engine)
```

### HAVING

- O comando `HAVING` permite aplicar um filtro ap√≥s fazer o `GROUP BY`

https://www.devmedia.com.br/exemplos-com-group-by-e-com-a-clausula-having-totalizando-dados-sql-server-2008-parte-2/19839

```python
query = '''
    SELECT product_category_name, COUNT(*) AS num_produtos
    FROM dbo.olist_products_dataset
    GROUP BY product_category_name
    HAVING COUNT(*) > 300
    ORDER BY COUNT(*) 
    
'''

pd.read_sql(query, engine)
```

#### Top 10 cidades com mais pedidos

```python
query = '''
    SELECT TOP 10 a.customer_city, COUNT(*) AS Orders
    FROM dbo.olist_customers_dataset AS a
    JOIN dbo.olist_orders_dataset AS b
    ON a.customer_id = b.customer_id
    GROUP BY a.customer_city
    ORDER BY COUNT(*) DESC
'''

pd.read_sql(query, engine)
```

#### Top 10 Categoria de produto com mais pedidos usando cart√£o de cr√©dito

```python
query = '''
    SELECT TOP 10 a.product_category_name, COUNT(*) as Num_pedidos
    FROM dbo.olist_products_dataset AS a
    
    JOIN dbo.olist_order_items_dataset AS b
    ON a.product_id = b.product_id
    
    JOIN dbo.olist_orders_dataset AS c
    ON b.order_id = c.order_id
    
    JOIN dbo.olist_order_payments_dataset AS d
    ON c.order_id = d.order_id
    
    WHERE d.payment_type = 'credit_card'
    GROUP BY a.product_category_name
    ORDER BY COUNT(*) DESC

'''

pd.read_sql(query, engine)
```

#### Top 10 Cidades com o maior valor de pagamento

```python
query = '''
    SELECT TOP 10 customer_city, MAX(a.payment_value) AS Highest_payment_value
    FROM dbo.olist_order_payments_dataset AS a
    
    JOIN dbo.olist_orders_dataset AS b
    ON a.order_id = b.order_id
    
    JOIN dbo.olist_customers_dataset AS c
    ON b.customer_id = c.customer_id
    
    GROUP BY c.customer_city
    ORDER BY Highest_payment_value DESC

'''

pd.read_sql(query, engine)
```

#### Top 10 vendedores com os maiores margens (pre√ßo - frete)

```python
query = '''
    SELECT TOP 10 seller_id
    , SUM(CAST(price AS FLOAT) - CAST(freight_value AS FLOAT)) AS margem
    FROM dbo.olist_order_items_dataset AS a
    GROUP BY seller_id
    ORDER BY margem DESC
'''

pd.read_sql(query, engine)
```

#### Top 10 vendedores com os piores margens

```python
query = '''
    SELECT TOP 10 seller_id
    , SUM(CAST(price AS FLOAT) - CAST(freight_value AS FLOAT)) AS margem
    FROM dbo.olist_order_items_dataset AS a
    GROUP BY seller_id
    ORDER BY margem
'''

pd.read_sql(query, engine)
```
